## Oppgave 1

### Oppgave 1a)
Vi skal løse initialverdiproblemet:
$$ \frac{d^2 y}{(d x)^2}  = -4 * \sin(2x), \quad y(0) = 0, \quad y'(0) = 2$$




Først flyttes dx^2 over på høyre side og så integreres uttrykket to ganger.


$$ \displaystyle\int{dy^2} = \int{-4*\sin(2x) dx^2} $$


$$dy = (4 * \frac{1}{2} \cos(2x) + C)dx$$


$$\displaystyle\int{dy} = \displaystyle\int{(2\cos(2x) + C)dx}$$


Da står vi igjen med dette uttrykket for y(x). 


$$ y(x) = sin(2x) + Cx + D$$


Og dette for y'(x).
$$ y'(x) = 2 \cos(2x) + C$$


Deretter bruker vi initialbetingelsene for å finne verdiene for C og D.
$$2 = 2 cos(0) + C \rightarrow C = 0 $$
$$ 0 = sin(0) + 0 + D \rightarrow D = 0 $$


Løsningen på initialverdiproblemet blir da:
$$ y(x) = sin(2x) $$


### Oppgave 1b)

Omskriv andreordens differensialligningen for skalaren y(x) i oppgave 1a) til en førsteordens differensialligning for vektoren $\vec{y}(x)$.


$$\vec{y}(x)  = \begin{pmatrix} y(x)\\ y'(x) \end{pmatrix} = \begin{pmatrix} sin(2x)\\ 2cos(2x) \end{pmatrix}$$


Og med initialbetingelsene er $\vec{y}(x)$:
$$\vec{y}(0)  = \begin{pmatrix} 0\\ 2 \end{pmatrix} $$

### Oppgave 1c)

```python
def f(x, y):
    return np.array([y[1], -4 * np.sin(2*x)])

def rk3 (x_start, x_slutt, y_0, h_0, tol, alpha, f):
   
    n, m = 0, 0                         # teller for antall godkjente steg (n) og ikke godkjente (m)
    x_verdier = np.array([x_start])     # array til x-verdier      
    y_verdier = np.array([y_0])         # array for y' og y
    h_verdier = np.array([])            # array til å lagre steglengde

    # Verdier som skal bli oppdatert i while-løkken
    x = x_start
    y = np.array(y_0)
    k1 = f(x,y)
    h = h_0
    
    while (x_slutt - x) > 0:

        # Velger steglengde
        h = min(h, x_slutt - x)

        k2 = f(x +h/2, y+ h*k1/2)
        k3 = f(x + 3*h/4, y + 3*h*k2/4)

        y_neste = y + h* (2*k1 + 3*k2 + 4*k3)/9             # y[n+1]

        k4 = f(x + h, y_neste) 

        z_neste = y + 1/24*h*(7*k1 + 6*k2 + 8*k3 + 3*k4)    # z[n+1]

        est = np.linalg.norm(y_neste - z_neste)             # || y[n+1] - z[n+1] ||
        
        if est < tol:       # Akseptert steg
            x += h
            y = y_neste
            k1 = k4
            n += 1
            # Lagrer steget og tilhørende verdier:  
            x_verdier = np.append(x_verdier, x)
            y_verdier = np.append(y_verdier, [y], axis=0)
            h_verdier = np.append(h_verdier, h)
        else:               
            m += 1          # Ikke akseptert steg

        # Regn ut h_ny og la h = h_ny:
        h_ny = alpha * h *(tol/est)**(1/3)
        h = h_ny
    
    return x_verdier, y_verdier, h_verdier, n, m
```

```python
# Verdier
x_start = 0
x_slutt = 2*np.pi
h_0 = np.pi / 100
y_0 = np.array([0, 2])
alpha = 0.8
tol = 1e-7

x_verdier, y_verdier, h_verdier, n, m = rk3(x_start, x_slutt, y_0, h_0, tol, alpha, f)

# Plot for y(x) og y'(x)
plt.plot(x_verdier, y_verdier[:,0], color = "cornflowerblue", label = "y(x)")
plt.plot(x_verdier, y_verdier[:,1], color = "plum", label = "y'(x)")

plt.title("Løsning for y(x) og y'(x) med RK3-metoden", size = 13)
plt.xlabel("x")     
plt.ylabel("y(x)")

# Endrer x-aksen til å ha pi
ticks = [0.5*np.pi, np.pi, 1.5*np.pi, 2*np.pi]
labels = [r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
plt.xticks(ticks, labels)

plt.tight_layout()
plt.grid(True)
plt.legend()
plt.show()
```

De analytiske løsningene av $y(x)$ og $y'(x)$ er (fra oppgave 1a): 

$ y(x) = \sin(2x)$ og $y'(x) = 2*\cos(2x)$

Den numeriske løsningen av $y(x)$ (blå graf) er en sinuskurve med amplitude 1, den går mellom [-1, 1]. Den oscillerer dobbelt så raskt som $\sin(x)$, dette er fordi argumentet $2x$ gir den en periode på $\pi$. Derfor ser vi to svingninger på intervallet $[0, 2\pi]$.
Grafen til $y'(x)$ (lilla) er en cosinuskurve med samme frekvens og dobbel amplitude av $y(x)$. Ekstremalpunktene til $y(x)$ svarer til nullpunktene i $y'(x)$.

Den numeriske løsningen stemmer overens med den analytiske, og RK3-metoden gir en nøyaktig tilnærming i dette tilfellet.

```python
# Plot for steglengde
plt.plot(x_verdier[:-1], h_verdier, label = "Steglengde", color = "hotpink")
plt.title("Steglengde i RK3-metoden", size = 13)
plt.ylabel("h(x)")
plt.xlabel("x")

# Endrer x-aksen til å ha pi
ticks = [0.5*np.pi, np.pi, 1.5*np.pi, 2*np.pi]
labels = [r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
plt.xticks(ticks, labels)

plt.tight_layout()
plt.grid(True)
plt.legend()
plt.show()
```

Frem til siste steget varierer steglengden $h(x)$ tilnærmet periodisk. Dette skyldes at løsningen endrer seg med varierende krummning. Når det er stor krummning, blir den lokale feilen større for en gitt $h$ og RK3-metoden må derfor redusere $h$ for å holde feilen under toleransen. I omeråder der løsningen er tilnærmet linær kan metoden bruke større steglengder.

RK3-metoden sjekker steglengden på hver iterasjon slik at $h$ ikke overskrider sluttpunktet $x_{slutt}$. I hvert steg settes derfor $h = min(h, x_{slutt} - x)$. 
I det siste steget vil dette føre til at $h$ blir lik avstanden til sluttpunktet. Denne steglengden vil avvike fra det tidligere mønsteret og grafen får derfor en "dupp" i siste steg.

$\alpha$ er en nedskaleringsfaktor. Når RK3-metoden beregner et nytt steg, bruker den feilen fra det forrige steget til å foreslå en ny steglengde $h_{ny}$. $\alpha$ gjør metoden mer konservativ, den gjør at RK3 - metoden bruker en mindre steglengde ennd et som er teoretisk optimalt, for å ikke feile på det neste steget. 

Plottet viser at for små verdier av $\alpha$ er metoden mer konservativ. Steglengden blir gjort litt mindre enn nødvendig, og algortimen bruker mange små steg. Da overlapper grafene bra, det er få steg som blir forkastet. 

Det er en synkende kruve mot $\alpha \approx 0.85 - 0.90$. Etterhvert som $\alpha$ øker bruker metoden større steglengde som resulterer i færre totalt antall steg. Ved $\alpha \approx 0.85 - 0.90$ når grafen sitt minimum. Dette er den optimale balansen mellom konservativ og aggressiv steglengdekontroll. 

Det er det ett oppsving ved $\alpha \approx 1.0$. Her overlapper ikke lenger de to grafene. Metoden tar for store steg som ikke tilfredsstiller toleransen, stegene forkastes og må gjentas. Det er disse forkastede stegene som gir oppsvinget i den lilla grafen. 

###     Oppgave 1e)

```python
def g(z):
    return z + np.sin(z) + np.cos(z)

def sekant_metoden (g, z0, z1, tol):
    z = [z0, z1]    # liste med startverdiene, z vil bli oppdatert med verdier som nærmer seg roten i while-løkken

    # første iterasjon, unngår indeksfeil i løkken (pga. abs(z[n-1] - z[n]))    
    n = 2
    z.append((z[n-2]*g(z[n-1]) - z[n-1]*g(z[n-2])) / (g(z[n-1]) - g(z[n-2])))

    # Itererer til forskjellen mellom z[n-1] - z[n] er mindre enn toleransen
    while abs(z[n-1] - z[n]) > tol:
        n+=1
        z.append((z[n-2]*g(z[n-1]) - z[n-1]*g(z[n-2])) / (g(z[n-1]) - g(z[n-2])))
        
    g_rot = z[-1]   # g_rot er z verdien som gir tilnærmet g(z) = 0
    return g_rot, z
```

```python
# Startverdier og toleranse
z_0 = 0
z_1 = 4
tol = 1e-10

# Finner roten til g(z) = z + sin(z) + cos(z) 
g_rot, _ = sekant_metoden (g, z_0, z_1, tol) 
print(f"Roten til g(z) = z + sin(z) + cos(z) er {g_rot:.5f}") 

# Verifisering
verdi = g(g_rot)
print(f"\nFor å verifisere at {g_rot:.5f} er roten til g(z), kan vi regne ut g(z):\ng({g_rot:.5f}) = {g(g_rot)}.\n\nAvrundet til 10 desimaler blir det: g({g_rot:.5f}) = {g(g_rot):.10f}")
```


Ved bruk av sekantmetoden er en tilnærmet rot av funksjonen $g(z) = z + \sin(z) + \cos(z)$ lik $-0.45662$ (avrundet til 5 desimaler).
Når denne verdien settes inn i funksjonen uten avrunding, får vi at $g(z) \approx 0$ så lenge vi avrunder med minst $15$ desimaler. Dette bekrefter at metoden har konvergert mot et nullpunkt.

Startverdiene og toleransen på $10^{-10}$ gir en svært nøyaktig måling på kort tid. 
Ved å variere toleransen ser vi at det er denne som i stor grad bestemmer nøyaktigheten til løsningen:

$tol = 10^{-4}$ gir avrunding til 0 frem til 10 desimaler.

$tol = 10^{-3}$ gir avrunding til 0 frem til 6 desimaler.

$tol = 10^{-2}$ gir avrunding til 0 frem til 3 desimaler.

Dette viser at en strengere toleranse gir høyere presisjon. Ved å endre på startverdiene ser vi at de har mindre innvirkning på sluttresultatet, så lenge de ligger i et intervall hvor metoden konvergerer. 

### Oppgave 1f)

```python
# Oppgitte startverdier
y_start = 0         # startpunkt [y(0) = 0]
y_slutt = 0         # endepunkt [y(2pi) = 0]
x_start = 0
x_slutt = 2*np.pi

# Parametere
b0 = 0          # b er startverdien y'(0) som gir yb(2pi) = 0
b1 = np.pi  
h_0 = np.pi / 100
alpha = 0.8
tol = 10e-7
# funksjonen f, er lik som i 1c) => gjenbruker samme

# 1.) Løser IVP for en gitt startverdi b = y'(0)
def løsning_IVP_med_b (b):
    y_0 = np.array([0, b])
    _, y_verdier_med_b, _, _, _ = rk3(x_start, x_slutt, y_0, h_0, tol, alpha, f)

    return y_verdier_med_b[-1, 0]   #sluttverdien y[2pi, b]

# 2.) Residualen F(b) = y[2pi, b] - y(2pi) 
def residual(b):
    return løsning_IVP_med_b(b) - y_slutt

# 3.) Finner hva b er med sekant_metoden fra 1e)
riktig_b, b_verdier_alle_iterasjoner = sekant_metoden(residual, b0, b1, tol)

# 4.) Løser IVP med RK3-metoden for hver b-verdi fra sekantmetoden
y_losninger = []
x_losninger = []

for b in b_verdier_alle_iterasjoner:

    ny_y_0 = np.array([0, b])
    x_verdier_midtid, y_verdier_midtid, _, _, _ = rk3(x_start, x_slutt, ny_y_0, h_0, tol, alpha, f)

    y_losninger.append(y_verdier_midtid)
    x_losninger.append(x_verdier_midtid)

# Plot
colors = ["midnightblue", "royalblue", "lightsteelblue", "slateblue" ]
for i in range(0, len(y_losninger)):
    plt.plot(x_losninger[i], y_losninger[i][:, 0], colors[i % len(colors)], label = f"Iterasjon nr. {i+1}")

plt.title("y(x) for hver iterasjon av sekantmetoden", size = 13)
plt.xlabel("x")     
plt.ylabel("y(x)")

# Endrer x-aksen til å ha pi
ticks = [0.5*np.pi, np.pi, 1.5*np.pi, 2*np.pi]
labels = [r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
plt.xticks(ticks, labels)

plt.tight_layout()
plt.grid(True)
plt.legend()
plt.show()
```

For hver iterasjon i sekantmetoden brukes $b = y'(0)$ som startverdi til å løse et IVP for $y(x)$ med RK3-metoden. Figuren viser løsningene $y(x)$ for hver iterasjon, i vårt tilfelle 4 iterasjoner. Målet er å finne b slik at randbetingelsen $y(2\pi) \approx 0$ oppfylles. Det er størst avvik i den første iterasjonen, mens iterasjon nummer $3$ og $4$ ender i   $0$. At kurvene for iterasjon $3$ og $4$ overlapper indikerer at sekantmetoden konvergerer svært raskt.

```python
# Oppgitte startverdier
y_start = 0         # startpunkt [y(0) = 0]
y_slutt = 0         # endepunkt [y(12) = 0]
x_start = 0
x_slutt = 12

# Parametere
b0 = 0          # b er verdien som gir yb(12) = 0
b1 = -4  
h_0 = 12 / 200
alpha = 0.8
tol = 10e-7

# IVP 
def IVP(x, y):
    return np.array([y[1], y[0] + np.sin(x)])

# 1.) Løser IVP for en gitt startverdi b = y'(0)
def løsning_IVP_med_b (b):
    y_0 = np.array([0, b])
    _, y_verdier_med_b, _, _, _ = rk3(x_start, x_slutt, y_0, h_0, tol, alpha, IVP)

    return y_verdier_med_b[-1, 0]   #sluttverdien y[12, b]

# 2.) Residualen F(b) = y[12, b] - y(12)
#     Funksjon fra 1f) residual(b)
def residual(b):
    return løsning_IVP_med_b(b) - y_slutt

# Starter timer (for å i 1h) kunne sammenlinge tiden)
timer_start_g = time.time()

# 3.) Finner hva b er med sekant_metoden fra 1e)
riktig_b, _ = sekant_metoden(residual, b0, b1, tol)

# 4.) Løser BVP med RK3 metoden og riktig_b
y_0_med_riktig_b = np.array([0, riktig_b])
x_verdier_g, y_verdier_g, h_g, n_g, m_g = rk3(x_start, x_slutt, y_0_med_riktig_b, h_0, tol, alpha, IVP)

# Avslutter timer 
timer_slutt_g = time.time()

# Plot for y(x)
plt.plot(x_verdier_g, y_verdier_g[:,0], color = "darkorange", label = "y(x)")

plt.title("Løsning av BVP: y'' = y + sin(x) på [0, 12]", size = 13)
plt.xlabel("x")    
plt.ylabel("y(x)")

plt.tight_layout()
plt.grid(True)
plt.legend()
plt.show()
```

Grafen viser løsningen av randverdiproblemet med skytemetoden kombinert med sekant-metoden. Vi ser at randverdiene $y(0) = y(12) = 0$ er oppfylt, noe som bekrefter at riktig initialderivert $y'(0)$ er funnet med rotløseren. Løsningen er glatt og oscillerende, noe som kan forventes på grunn av $\sin(x)$. 

### Oppgave 1h)

```python
from scipy.integrate import solve_bvp

# Grensebetingelser: y(0) = y(12) = 0
def bc(ya, yb):
    return [ya[0]-0, yb[0]-0]

# Parametere, samme verdier som 1g)
# x_verdier likt som i 1g)
y_init = np.zeros((2, x_verdier_g.size))

# Starter timer (for å kunne senere sammenlinge med 1g)
timer_start_h = time.time()

# scipy.integrate.solve_bvp
sol = solve_bvp (IVP, bc, x_verdier_g , y_init)

# y-verdier
y_verdier_h = sol.sol(x_verdier_g)[0]

# Avslutter timer
timer_slutt_h = time.time()

# Plot for y(x) med solve_bvp
plt.plot(x_verdier_g, y_verdier_h, label = 'Numerisk løsning', color = 'chocolate')

plt.title("Løsning av y'' = y + sin(x) med solve_bvp", size = 13)
plt.xlabel("x")    
plt.ylabel("y(x)")

plt.tight_layout()
plt.grid(True)
plt.legend()
plt.show()
```

Plottet viser samme funksjon som i 1g) men løst med scipy sin numeriske løser solve_bvp. Vi ser at randbetingelsene fortsatt tilfredstilles og grafen har en sinuslignende oppførsel (som skyldes sin(x)-leddet i diffrensiallikningen).

```python
# Sammenlinger 
differanse = []

for i in range(0, len(x_verdier_g)): 
    diff = abs(y_verdier_h[i] - y_verdier_g[:,0][i])
    differanse.append(diff)

# Plot for differanse mellom sekant+BVP og solve_bvp
plt.plot(x_verdier_g, differanse, label = "Differanse", color = "rebeccapurple")

plt.title("Differanse mellom egen løser og solve_bvp", size = 13)
plt.xlabel("x")    
plt.ylabel("|y_egen_solver - y_solve_bvp|")

plt.tight_layout()
plt.grid(True)
plt.legend()
plt.show()
```

Grafen viser differansen i hvert punkt på x-aksen mellom 1g) og 1h), $\left| y_h(x) - y_g(x) \right|$. Differansen er svært liten på hele intervallet, y-aksen har en størrelsesorden på $10^{-6}$.

Ved start- og ekstremalpunktene er differansen nær 0, løsningene er tilnærmet like i disse områdene. Generelt viser plottet at de numeriske metodene brukt for å løse 1g) og 1h) er gir tilnærmet identiske resultater.

```python
def matrices_to_vector(matrix1, matrix2, matrix3, matrix4):
    v1 = complex_matrix_to_real_vector(matrix1) #først gjør jeg alle matrisene til en lang vektor ved hjelp av tidligere funksjon
    v2 = complex_matrix_to_real_vector(matrix2)
    v3 = complex_matrix_to_real_vector(matrix3)
    v4 = complex_matrix_to_real_vector(matrix4)

    long_vector = make_long_vector(v1,v2,v3,v4) # så slår jeg alle sammen sammen
    return long_vector

def vector_to_matrices(vector):

    v1,v2,v3,v4 = split_vector(vector)

    m1 = one_D_vector_to_complex_matrix(v1)
    m2 = one_D_vector_to_complex_matrix(v2)
    m3= one_D_vector_to_complex_matrix(v3)
    m4 = one_D_vector_to_complex_matrix(v4)

    return m1, m2, m3, m4
```

## Oppgave 2d

```python
def differantiate_vector(vector, e):

    gamma, gamma_tilde, omega, omega_tilde = vector_to_matrices(vector) # Splitter opp til matriser

    I2 = np.eye(2, dtype=complex)
    N_tilde = np.linalg.solve(I2 - gamma_tilde @ gamma, I2) # Fra oppgave
    N = np.linalg.solve(I2 - gamma @ gamma_tilde, I2)

    d_dx_gamma = omega
    d_dx_gamma_tilde = omega_tilde
    d_dx_omega = -2j*(e+0.01j)*gamma - 2*omega@N_tilde@gamma_tilde@omega
    d_dx_omega_tilde = -2j*(e+0.01j)*gamma_tilde - 2*omega_tilde@N@gamma@omega_tilde

    differentiated = matrices_to_vector(d_dx_gamma,d_dx_gamma_tilde,d_dx_omega,d_dx_omega_tilde) # Gjør tilbake til vector

    return differentiated
```

### Oppgave 2e)

```python
def function_to_solve(x_vector, vec_matrix, e ): # På formen som solve_bvp ønsker seg
    size = vec_matrix.shape

    differantiated_matrix = np.empty(size) # Lager en tom matrise

    for index, column in enumerate(vec_matrix.T):
        differentiated = differantiate_vector(column, e) # Deriverer matrisen
        differantiated_matrix[:,index] = differentiated # Fyller inn verdier i matrisen

    return differantiated_matrix
```

### Oppgave 2f)

```python
def lhs (v_left, v_right, l, e):

    # Randverdiene på venstre kant
    gamma_0, gamma_thilde_0, omega_0, omega_thilde_0 = vector_to_matrices(v_left) # Splitter opp vektoren

    # Randverdiene på høyre kant
    gamma_l, gamma_thilde_l, omega_l, omega_thilde_l = vector_to_matrices(v_right) # Splitter opp vektoren

    zeta = 3

    likning_13 = omega_0 - gamma_0 / (zeta * l) # Likninger fra oppgave
    likning_14 = omega_thilde_0 - gamma_thilde_0 / (zeta * l)
    likning_15 = omega_l + gamma_l / (zeta * l)
    likning_16 = omega_thilde_l + gamma_thilde_l / (zeta * l)

    return matrices_to_vector(likning_13, likning_14, likning_15, likning_16) # Tilbake til vektor
```

### Oppgave 2h)

```python
def calculate_density_of_states(function_to_solve, bc, x, y, epsilon, l, midtverdi = False):
    sol = solve_bvp(lambda x, y: function_to_solve(x_vector= x, vec_matrix= y, e = epsilon),lambda x, y: bc(v_left = x, v_right = y, l= l, e = epsilon), x, y)
    solution = sol.sol(x)

    I2 = np.eye(2, dtype=complex) # Identitetsmatrisen

    rho = np.block([
            [np.identity(2),        np.zeros((2,2))],
            [np.zeros((2,2)),   -np.identity(2)]])

    if midtverdi:
        # Vi trenger kun å se på midten, ref. "plot the density in the middle of the metal"
        mid = len(x) // 2
        gamma, gamma_tilde, omega, omega_tilde = vector_to_matrices(solution[:,mid])

        N_tilde = np.linalg.solve(I2 - gamma_tilde @ gamma, I2)
        N = np.linalg.solve(I2 - gamma @ gamma_tilde, I2)

        green_function = np.block([
            [2*N - np.identity(2),        2*N @ gamma],
            [-2*N_tilde @ gamma_tilde,   -2*N_tilde + np.identity(2)]])

        D_mid = 1/4 * np.real(np.trace(rho@green_function)) # Regner ut Dos for midten

        return D_mid, solution

    dos_list = []
    for i in range(len(x)):

        gamma, gamma_tilde, omega, omega_tilde = vector_to_matrices(solution[:,i])

        N_tilde = np.linalg.solve(I2 - gamma_tilde @ gamma, I2)
        N = np.linalg.solve(I2 - gamma @ gamma_tilde, I2)

        green_function = np.block([
        [2*N - np.identity(2),        2*N @ gamma],
        [-2*N_tilde @ gamma_tilde,   -2*N_tilde + np.identity(2)]])

        Density_of_states_per_Do = 1/4 * np.real(np.trace(rho@green_function))
        dos_list.append(Density_of_states_per_Do)
    
    return dos_list, solution
```

### Oppgave 2i)

```python
#Gjentagelse av oppgave 2f) men denne gangen antar vi at det er to superledere på hver side av metallet

#uttrykk brukt i Riccati matrisene
def vartheta_sigma(sigma, e):
    
    return np.atanh(sigma / (e + 0.01 *1j))

def s_sigma(sigma, e):
    return np.sinh(vartheta_sigma(sigma, e))

def c_sigma(sigma, e):
    return np.cosh(vartheta_sigma(sigma, e))

#Riccati matrisene
def gamma_L( phi_L, e):
    return np.exp(phi_L *1j)*np.block([
        [0,        s_sigma(1, e)/(1+c_sigma(1, e))],
        [s_sigma(-1, e)/(1+c_sigma(-1, e)),   0]])

def gamma_L_tilde(phi_L, e):
    return np.exp(-phi_L * 1j)*np.block([
        [0,        s_sigma(-1, e)/(1+c_sigma(-1, e))],
        [s_sigma(1, e)/(1+c_sigma(1, e)),   0]])

def gamma_R(phi_R, e):
    return np.exp(phi_R * 1j)*np.block([
        [0,        s_sigma(1, e)/(1+c_sigma(1, e))],
        [s_sigma(-1, e)/(1+c_sigma(-1, e)),   0]])

def gamma_R_tilde( phi_R, e):
    return np.exp(-phi_R *1j)*np.block([
        [0,        s_sigma(-1, e)/(1+c_sigma(-1, e))],
        [s_sigma(1, e)/(1+c_sigma(1, e)),   0]])
    

def bc_superleder(v_left, v_right, l, e):
    global phi_L
    global phi_R
    # Randverdiene på venstre kant
    gamma_0, gamma_tilde_0, omega_0, omega_tilde_0 = vector_to_matrices(v_left)

    # Randverdiene på høyre kant
    gamma_l, gamma_tilde_l, omega_l, omega_tilde_l = vector_to_matrices(v_right)
    
    zeta = 3

    I2 = np.eye(2, dtype=complex) # Identitetsmatrisen

    gL = gamma_L(phi_L, e)
    gtL = gamma_L_tilde(phi_L, e)

    gR = gamma_R(phi_R, e)
    gtR = gamma_R_tilde(phi_R, e)

    N_L = np.linalg.solve(I2 - gL @ gtL, I2)
    N_L_tilde = np.linalg.solve(I2 - gtL @ gL, I2)

    N_R = np.linalg.solve(I2 - gR @ gtR, I2)
    N_R_tilde = np.linalg.solve(I2 - gtR @ gR, I2)

    likning_13 = omega_0 + 1/(zeta*l) * (np.identity(2)-gamma_0@gamma_L_tilde(phi_L, e)) @ N_L @(gamma_L(phi_L, e)-gamma_0)
    likning_14 = omega_tilde_0 + 1/(zeta*l) * (np.identity(2)-gamma_tilde_0@gamma_L(phi_L, e)) @ N_L_tilde @(gamma_L_tilde(phi_L, e)-gamma_tilde_0)

    likning_15 = omega_l - 1/(zeta*l) * (np.identity(2)-gamma_l@gamma_R_tilde(phi_R, e)) @ N_R @(gamma_R(phi_R, e)-gamma_l)
    likning_16 = omega_tilde_l - 1/(zeta*l) * (np.identity(2)-gamma_tilde_l@gamma_R(phi_R, e)) @ N_R_tilde @(gamma_R_tilde(phi_R, e)-gamma_tilde_l)

    return matrices_to_vector(likning_13, likning_14, likning_15, likning_16)


```

```python
def calculate_intergrand(y_init, l, e, x, midten = False):

    sol = solve_bvp(lambda x, y: function_to_solve(x_vector= x, vec_matrix= y, e = e),lambda x, y:  bc_superleder(v_left = x, v_right = y, l= l, e= e), x, y_init)
    solution = sol.sol(x)

    current_list = []

    I = np.identity(2)
    Z = np.zeros((2,2))
    rho = np.block([[I, Z], [Z, -I]])

    if midten:
        gamma, gamma_tilde, omega, omega_tilde = vector_to_matrices(solution[:, len(x)//2]) # Splitter vektor

        N_tilde = np.linalg.solve(I - gamma_tilde@gamma, I)
        N = np.linalg.solve(I- gamma@gamma_tilde, I)

        green_function = np.block([ # Green funksjonen
            [2*N - I ,        2*N @ gamma],
            [-2*N_tilde @ gamma_tilde,   -2*N_tilde + I]])
        
        d_N_dx = N @ (omega @ gamma_tilde + gamma @ omega_tilde)@ N # Den deriverte av N mhp x

        d_N_tilde_dx = N_tilde @ (omega_tilde @ gamma + gamma_tilde @ omega) @ N_tilde # Den deriverte av N tilde mhp x

        d_green_dx = 2*np.block([
        [d_N_dx, N@ omega + (d_N_dx)@ gamma],
        [-N_tilde @ omega_tilde - (d_N_tilde_dx @ gamma_tilde), - d_N_tilde_dx]]) # Den deriverte av green funksjonen mhp x

        current = np.real(np.trace(rho@(green_function @ d_green_dx - d_green_dx @ green_function)))

        return solution, current

    for i in range(len(x)):
        gamma, gamma_tilde, omega, omega_tilde = vector_to_matrices(solution[:, i])

        N_tilde = np.linalg.solve(I - gamma_tilde@gamma, I)
        N = np.linalg.solve(I- gamma@gamma_tilde, I)

        green_function = np.block([
            [2*N - I ,        2*N @ gamma],
            [-2*N_tilde @ gamma_tilde,   -2*N_tilde + I]]) # Green funksjonen
        
        d_N_dx = N @ (omega @ gamma_tilde + gamma @ omega_tilde)@ N # Den deriverte av N mhp x

        d_N_tilde_dx = N_tilde @ (omega_tilde @ gamma + gamma_tilde @ omega) @ N_tilde # Den deriverte av N tilde mhp x

        d_green_dx = 2*np.block([
        [d_N_dx, N@ omega + (d_N_dx)@ gamma],
        [-N_tilde @ omega_tilde - (d_N_tilde_dx @ gamma_tilde), - d_N_tilde_dx]]) # Den deriverte av green funksjonen mhp x

        current = np.real(np.trace(rho@(green_function @ d_green_dx - d_green_dx @ green_function)))
        current_list.append(current)

    return solution, current_list
```

```python
l = 1
m = 101
x = np.linspace(0, l, m)
epsilon = [2,1.5, 1, 0.5,0]
color_list = ["cornflowerblue", "plum", "hotpink", "mediumseagreen", "salmon"]

y_init = np.zeros((32, m))

for index, e in enumerate(epsilon):
    solution, current_list = calculate_intergrand(y_init, l, e, x)
    y_init = solution # Bruker den forrige løsningen som nytt initialgjett
    
    plt.plot(x, current_list, color = color_list[index], label=f"$\\varepsilon = {e}$")

plt.xlabel("Position $x$")
plt.ylabel(r"$j(x, \varepsilon)$")
plt.title("Current intergrand as function of position")

plt.grid()
plt.legend()
plt.show()
```
